```python
#處理xml檔 -> 沒有現成的函數將xml讀入pandas DataFrame中 -> 自己開發
import pandas
import xml.etree.ElementTree as ET
tree = ET.parse('myxml.xml')  #ElementTree is not iterable
print(tree)

root = tree.getroot()  #取得根結點，方便之後的操作
print(root)


#print出每一個element中的內容
for child in root:
    print(child.tag, child.attrib)
    inner_name = child.get('name')  #取得屬性對應的值
    print(inner_name)
    
    #找尋子結點

#用index找出要的node
print(root[0][2].text)


#將xml檔轉換成pandas的DataFrame
columns = ['name','rank','gdppc']
dataframe = pandas.DataFrame(columns = columns)
for country in root:
    name = country.attrib.get('name')
    rank = country.find('rank').text if country is not None else None
    gdppc = country.find('gdppc').text if country is not None else None
    
    #Series的index對應每個column
    #Series是一行資料(column) -> 填入dataframe時，轉置成row, 每個index對應dataframe中的column
    #dataframe的index對應每個row data
    dataframe = dataframe.append(pandas.Series([name,rank,gdppc], index = columns), ignore_index = True)  #Series: 看來是表格中的一行資料(column)
    
print(dataframe)
```

```python
#排除離群值 -> 自行定義function
import numpy as np
import pandas as pd

outliers = []  #紀錄離群值的list


#找出與平均值之差超過3個標準差的值
def detect_outlier(data):
    #data是pandas的dataframe
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)
    
    for y in data:
        z_score = (y - mean) / std  #標準分數
        if np.abs(z_score) > threshold:
            outliers.append(y)
    
    return outliers


#my_data = [60,65,70,55,59,78,80,92,95,80,77,100,1000,10,47]
#my_outliers = detect_outlier(my_data)
#print(my_outliers)


push_pay_list = pd.read_excel('1120309燃料費催繳清冊1061231以前__增加排除條件.xlsx', sheet_name = '106年以前燃料費催繳清冊__增加排除狀態')
print(push_pay_list)
#print(outliers)
print('---------------------------------')
outlier_push_pay = detect_outlier(push_pay_list['應納金額'])
#print(outlier_push_pay)
#print(outliers)
print('---------------------------------')




#濾出超過3個標準差的資料(以「應納金額」欄位值做運算)
def cal_zscore(data):
    #data是某個要判斷的list
    zscore = []
    mean = np.mean(data)
    std = np.std(data)
    
    print(mean)
    print(std)
    return (data - mean) / std

#print(cal_zscore(push_pay_list['應納金額']))
print('---------------------------------')

print(type(push_pay_list['應納金額']))
print(push_pay_list['應納金額'])
z_scores = cal_zscore(push_pay_list['應納金額'])
abs_z_scores = np.abs(z_scores)
test = (abs_z_scores > 3)  #會是一個True與False組成的series
print(type(test))
test1 = [True] * 37352
print(test)

print('---------------------------------')

filtered_push_pay_list = push_pay_list[abs_z_scores > 3]  #看起來dataframe中的filter([]中)只在乎True或False, 只要是True的就會留下
print(filtered_push_pay_list)
print(push_pay_list[test1])
                      
                      
```

```python
#資料的合併
import pandas as pd

#dataframe1 and 2
dataframe1 = pd.DataFrame({'lkey':['foo','bar','baz','foo'], 'lseq':['0001','0002','0003','0004'], 'value':[1,2,3,4]})
dataframe2 = pd.DataFrame({'rkey':['foo','bar','qux','bar','tes'], 'rseq':['0050','0002','0313','4105','5510'], 'value':[1,6,7,5,10]})
print(dataframe1)
print('----------------------')
print(dataframe2)
print('----------------------')

#merge -> left join
dataframe1.merge(dataframe2, left_on = ['lkey','lseq'], right_on = ['rkey','rseq'], how = 'left')  #on的順序要一樣!!


#join -> 特別的操作，不同於database中table的join @@
#屬於index-based
#相同index對接(同一個row的資料相連), column - index對接 -> 很奇怪
#橫向發展
print(dataframe1.join(dataframe2, lsuffix = '_left', rsuffix = '_right', how = 'right') )


#concat -> 跟join一樣，index-based, (不知所以然)
#可橫可縱向發展 -> 縱向發展就如同union all(前提: column name一致)
#單純地把資料接起來，不像merge能夠根據一樣的內容來連接
frames = [dataframe1, dataframe2]
pd.concat(frames, keys=['a','b'], axis = 0, join = 'inner')  #keys: 階層式的索引; axis: 0 index / 1 column; join: 要變瘦or變胖: inner 變瘦 / outer 變胖


print('----------------------')
#stack的用法 -> 把column轉成row
df_stack = dataframe1.stack()
print(df_stack)
print('----------------------')

#row轉成column
df_unstack = df_stack.unstack()  
print(df_unstack)
```
